# Gráfico de Barras

## Importação das Bibliotecas

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

## Leitura dos dados

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

## Plotagem do Gráfico

```python
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000) #Simplifica o número
sns.set_style('whitegrid')
plt.figure(figsize=(5,3)) 
ax = sns.barplot(data=df, y='titulo', x='arrecadacao_milhao', color='tomato', alpha=0.95)

ax.bar_label(
    ax.containers[0], 
    fmt='$%.1f', 
    label_type='center',
    color='white',
)

plt.xlabel('Arrecadação (Milhões de Dolares)')
plt.ylabel('Filme')
plt.title('Arrecadação de Filmes')

ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)

plt.show()
```
