```python
df_dict = pd.DataFrame([
{
"variavel":"species",
"descricao": "Especie do pinguin",
"tipo":"qualitativa",
"subtipo":"nominal",
},{
"variavel":"island",
"descricao":"Ilha onde vive o pinguin",
"tipo":"qualitativa",
"subtipo":"nominal",
},{
"variavel":"bill_length_mm",
"descricao":"Comprimento do bico do pinguin em milimetros",
"tipo":"quantitativa",
"subtipo":"continua",
},{
"variavel":"bill_depth_mm",
"descricao":"Profundidade do bico do pinguin em milimetros",
"tipo":"quantitativa",
"subtipo":"continua",
},{
"variavel":"flipper_length_mm",
"descricao":"Comprimento da nadadeira do pinguin em milimetros",
"tipo":"quantitativa",
"subtipo":"continua",
},{
"variavel":"body_mass_g",
"descricao":"Peso do pinguin em gramas",
"tipo":"quantitativa",
"subtipo":"continua",
},{
"variavel":"sex",
"descricao":"Sexo do pinguin",
"tipo":"qualitativa",
"subtipo":"nominal",
}
])
df_dict
```

### Estatisticas qualitativas

```python
df[['species','island','sex']].mode()
```

### Estatisticas quantitativas

```python
df.describe()
```

```python
qualitative_vars = ['species','island','sex']

for variavel in qualitative_vars:
	valores_unicos = df[variavel].unique()
	for valor in valores_unicos:
		print(f"{variavel} - {valor}")
		subamostra = df[df[variavel] == valor]
		display(subamostra.describe())
		print("-"*80)
```
